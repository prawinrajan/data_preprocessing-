```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv('/home/prawin/final_heart_data.csv')
```

```python
df.tail()
```

```python
def replace_age(val):
    if val < 55:
        return 0
    else:
        return 1
df['age']=df['age'].apply(replace_age,1)
```

```python
def replace_trestbps(val):
    if val < 140:
        return 0
    else:
        return 1
df['trestbps']=df['trestbps'].apply(replace_trestbps,1)
```

```python
def replace_chol(val):
    if val < 240:
        return 0
    else:
        return 1
df['chol']=df['chol'].apply(replace_chol,1)
```

```python
def replace_thalach(val):
    if val < 165:
        return 0
    else:
        return 1
df['thalach']=df['thalach'].apply(replace_thalach,1)
```

```python
def replace_target(val):
    if val == 0:
        return 0
    else:
        return 1
df['num']=df['num'].apply(replace_target,1)
```

```python
df.tail()
```

```python
df['thal'].unique()
```

```python
#replace NULL value by mean values

df['ca'].fillna(df['ca'].mean(),inplace=True)

df['thal'].fillna(df['thal'].mean(),inplace=True)
df.tail()
```

```python
def replace_thalach(val):
    if val < 2:
        return 0
    elif val <= 1.5 and val >=4.2 :
        return 1
    elif val > 2.5:
        return 2
df['oldpeak']=df['oldpeak'].apply(replace_thalach,1)
```

```python
df =pd.read_csv('/home/prawin/heart_data.csv')
df.tail()
```

```python
a = pd.get_dummies(df['cp'], prefix = "cp")
b = pd.get_dummies(df['thal'], prefix = "thal")
c = pd.get_dummies(df['slope'], prefix = "slope")
d = pd.get_dummies(df['ca'], prefix = "ca")
e=pd.get_dummies(df['restecg'], prefix = "restecg")
```

```python
frames = [df, a, b, c,d,e]
df = pd.concat(frames, axis = 1)
df.head()
```

```python
df = df.drop(columns = ['cp', 'thal', 'slope','ca','restecg'])
df.head()
```

```python
df.to_csv('final_heart_data.csv') 
```
